```python
# Import libraries
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
```

```python
#AND truth table   as Dataset
data={'X1':[0,0,1,1],
      'X2':[0,1,0,1],
      'Y':[0,1,1,0] 
        }
df=pd.DataFrame(data)
df
```

```python
# spliting data into input and output class
X=df.iloc[:,0:2].values
Y=df.iloc[:,2].values
X,Y
```

```python
# Model with 2 input neurons, 1 hidden layers with 4 neurons and 1 output layer
model = Sequential()
model.add(Dense(4, input_dim=2, activation='sigmoid',name="Hidden"))
model.add(Dense(1, activation='sigmoid',name="output"))
```

```python
model.layers
```

```python
model.summary()
```

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on the dataset
History=model.fit(X, Y,epochs=5000)
```

```python
History.params
```

```python
History.history.keys()
```

```python
#History.history['loss']
```

```python
pred=model.predict(X)
pred=np.where(pred>0.5,1,0)
pred
```

```python
# Generate generalization metrics
score = model.evaluate(X, Y, verbose=1)
print(f'Test loss: {score[0]} / Test accuracy: {score[1]}')
```

```python
# Evaluating the performance 
from matplotlib import pyplot as plt 
epochs = range(0,5000)
plt.plot(epochs, History.history['loss'], 'g', label='Training Loss')
plt.title('Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Evaluating the performance 
from matplotlib import pyplot as plt 
epochs = range(0,5000)
plt.plot(epochs, History.history['accuracy'], 'g', label='Training accuracy')
plt.title('Training accuracy')
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```
